```python
import os
from pathlib import Path
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from datasets import load_from_disk
import torch
from torchvision.transforms import ToPILImage

# Set paths directly in the code
test_dataset_path = "data/vgg_train"  # Replace with your dataset path

def load_test_dataset(test_dataset_path):
    if os.path.exists(test_dataset_path):
        dataset = load_from_disk(test_dataset_path)["train"]
    else:
        raise FileNotFoundError(f"Dataset not found at {test_dataset_path}")
    return dataset

def show_spectrograms(dataset):
    for idx, example in enumerate(dataset):
        if idx <= 4:
            # Get the image tensor from the dataset
            image = example["image"]
            
            # Display the image
            plt.figure(figsize=(5, 5))
            plt.imshow(image, cmap="gray")
            plt.title(f"Spectrogram {idx}")
            plt.axis("off")
            plt.show()

# Load the dataset
test_dataset = load_test_dataset(test_dataset_path)

# Show the spectrograms
show_spectrograms(test_dataset)
```
